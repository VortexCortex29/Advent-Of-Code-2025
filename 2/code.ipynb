```python
import math
from bisect import bisect_left, bisect_right
```

```python
def parse_input(file_name):
    with open(file_name) as f:
        data = f.read()
    data = data.split(",")
    data = [list(map(int, x.split("-"))) for x in data]
    return data
input = parse_input("input.txt")
example = parse_input("example.txt")
```

```python
def part_1(input):
    largest = max(input, key=lambda x: x[1])[1]
    d = int(math.log10(largest) + 1)
    invalid_ids = []
    for i in range(1, 10**(d//2)):
        invalid_ids.append(int(str(i) + str(i)))

    prefix_sum = [0]
    for i in invalid_ids:
        prefix_sum.append(prefix_sum[-1] + i)

    total = 0
    for a, b in input:
        l = bisect_left(invalid_ids, a)
        r = bisect_right(invalid_ids, b)
        total += prefix_sum[r] - prefix_sum[l]
    return total
```

```python
assert part_1(example) == 1227775554
print(part_1(input))
```

```python
def part_2(input):
    largest = max(input, key=lambda x: x[1])[1]
    d = int(math.log10(largest) + 1)
    invalid_ids = set()
    for i in range(1, 10**(d//2)):
        s = str(i) + str(i)
        while int(s) <= largest:
            invalid_ids.add(int(s))
            s += str(i)
    invalid_ids = sorted(invalid_ids)

    prefix_sum = [0]
    for i in invalid_ids:
        prefix_sum.append(prefix_sum[-1] + i)

    total = 0
    for a, b in input:
        l = bisect_left(invalid_ids, a)
        r = bisect_right(invalid_ids, b)
        total += prefix_sum[r] - prefix_sum[l]
    return total
```

```python
assert part_2(example) == 4174379265
print(part_2(input))
```
