```python
import math
```

```python
def parse_input(file_name):
    with open(file_name) as f:
        data = f.read()
    data = data.split("\n")
    return data
input = parse_input("input.txt")
example = parse_input("example.txt")
```

```python
def part_1(input):
    input = [d.split() for d in input]
    for i in range(len(input) - 1):
        for j in range(len(input[i])):
            input[i][j] = int(input[i][j])
    n, m = len(input), len(input[0])
    total = 0
    for j in range(m):
        curr = 0
        if input[-1][j] == '*':
            curr = 1
        for i in range(n - 1):
            if input[-1][j] == '*':
                curr *= input[i][j]
            else:
                curr += input[i][j]
        total += curr
    return total
```

```python
assert(part_1(example) == 4277556)
print(part_1(input))
```

```python
def part_2(input):
    n, m = len(input), len(input[0])
    buffer = []
    temp = []
    for j in range(m - 1, -1, -1):
        s = ""
        for i in range(n - 1):
            s += input[i][j]
        s = s.split()
        if s:
            s = s[0]
            temp.append(int(s))
            if input[-1][j] in ['+', '*']:
                buffer.append(temp)
                temp = []
    ops = input[-1].split()
    total = 0
    for i, op in enumerate(ops[::-1]):
        if op == '+':
            total += sum(buffer[i])
        if op == '*':
            total += math.prod(buffer[i])
    return total
```

```python
assert part_2(example) == 3263827
print(part_2(input))
```
