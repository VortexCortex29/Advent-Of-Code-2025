```python
def parse_input(file_name):
    with open(file_name) as f:
        data = f.read()
    data = list(map(list, data.split("\n")))
    return data
input = parse_input("input.txt")
example = parse_input("example.txt")
```

```python
def part_1(input):
    n, m = len(input), len(input[0])
    beams = set([input[0].index('S')])
    splits = 0
    for i in range(1, n):
        for j in range(m):
            if j in beams:
                if input[i][j] == '^':
                    beams.remove(j)
                    beams.add(j - 1)
                    beams.add(j + 1)
                    splits += 1
    return splits
```

```python
assert part_1(example) == 21
print(part_1(input))
```

```python
def part_2(input):
    n, m = len(input), len(input[0])
    beams = set([input[0].index('S')])
    for i in range(1, n):
        for j in range(m):
            if j in beams:
                if input[i][j] == '^':
                    beams.remove(j)
                    beams.add(j - 1)
                    beams.add(j + 1)
                    input[i][j + 1], input[i][j - 1] = '|', '|'
                else:
                    input[i][j] = '|'
    dp = [0] * m
    for j in range(m):
        if input[n - 1][j] == '|':
            dp[j] = 1
    for i in range(n - 2, -1, -1):
        new_dp = [0] * m
        for j in range(m):
            if input[i][j] == 'S':
                return dp[j]
            if input[i][j] == '|':
                new_dp[j] = dp[j]
            if input[i][j] == '^':
                new_dp[j] = dp[j - 1] + dp[j + 1]
        dp = new_dp
```

```python
assert part_2(example) == 40
print(part_2(input))
```
