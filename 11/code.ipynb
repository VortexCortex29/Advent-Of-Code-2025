```python
from collections import defaultdict, deque
```

```python
def parse_input(file_name):
    with open(file_name) as f:
        data = f.read()
    data = data.split("\n")
    data = [(d.split(":")[0], d.split(":")[1]) for d in data]
    data = {src: dst.split() for src, dst in data}
    return data
input = parse_input("input.txt")
example_1 = parse_input("example_1.txt")
example_2 = parse_input("example_2.txt")
```

```python
def part_1(input):
    q = deque(['you'])
    paths = 0
    while q:
        node = q.popleft()
        if node == 'out':
            paths += 1
        else:
            q += input[node]
    return paths
```

```python
assert(part_1(example_1) == 5)
print(part_1(input))
```

```python
def part_2(input):
    in_degree = {k: 0 for k in input.keys()}
    in_degree["out"] = 0
    for nodes in input.values():
        for node in nodes:
            in_degree[node] += 1
    q = deque([node for node in input.keys() if in_degree[node] == 0])
    topo = []
    while q:
        node = q.popleft()
        topo.append(node)
        if node == 'out':
            break
        for nei in input[node]:
            in_degree[nei] -= 1
            if in_degree[nei] == 0:
                q.append(nei)
    def count_paths(start, end):
        paths = defaultdict(int)
        paths[start] = 1
        for node in topo:
            if node == end:
                return paths[end]
            for nei in input[node]:
                paths[nei] += paths[node]
        return 0
    
    return count_paths('svr', 'fft') * count_paths('fft', 'dac') * count_paths('dac', 'out') + count_paths('svr', 'dac') * count_paths('dac', 'fft') * count_paths('fft', 'out')
```

```python
assert(part_2(example_2) == 2)
print(part_2(input))
```
