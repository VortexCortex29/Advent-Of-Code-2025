```python
def parse_input(file_name):
    with open(file_name) as f:
        data = f.read()
    data = data.split("\n")
    return data
input = parse_input("input.txt")
example = parse_input("example.txt")
```

```python
def part_1(input):
    total = 0
    for line in input:
        best = 0
        for i in range(len(line)):
            for j in range(i + 1, len(line)):
                best = max(best, int(line[i] + line[j]))
        total += best
    return total
```

```python
assert part_1(example) == 357
print(part_1(input))
```

```python
def part_2(input):
    total = 0
    for line in input:
        stack = []
        for i, l in enumerate(line):
            while stack and stack[-1] < l and len(stack) - 1 + len(line) - i >= 12:
                stack.pop()
            if len(stack) < 12:
                stack.append(l)
        total += int("".join(stack))
    return total
```

```python
assert part_2(example) == 3121910778619
print(part_2(input))
```
