```python
def parse_input(file_name):
    with open(file_name) as f:
        data = f.read()
    data = data.split("\n\n")
    data = [x.split("\n") for x in data][0]
    return data
input = parse_input("input.txt")
example = parse_input("example.txt")
```

```python
def part_1(input):
    lock = 50
    password = 0
    for line in input:
        dir, rot = line[0], int(str(line[1:]))
        if dir == 'L': rot = -rot
        lock = (lock + rot) % 100
        if lock == 0:
            password += 1

    return password
```

```python
assert part_1(example) == 3
print(part_1(input))
```

```python
def part_2(input):
    lock = 50
    password = 0
    for line in input:
        dir, rot = line[0], int(str(line[1:]))
        if dir == 'L':
            password += ((100 - lock) % 100 + rot) // 100
            lock = (lock - rot) % 100
        else:
            password += (lock + rot) // 100
            lock = (lock + rot) % 100

    return password
```

```python
assert part_2(example) == 6
print(part_2(input))
```
