```python
def parse_input(file_name):
    with open(file_name) as f:
        data = f.read()
    data = data.split("\n")
    idx = data.index('')
    intervals, ids = data[:idx], data[idx + 1:]
    ids = [int(i) for i in ids]
    intervals = [[int(a), int(b)] for a, b in (s.split('-') for s in intervals)]
    return intervals, ids
input = parse_input("input.txt")
example = parse_input("example.txt")
```

```python
def part_1(input):
    intervals, ids = input
    count = 0
    for id in ids:
        for a, b in intervals:
            if a <= id <= b:
                count += 1
                break
    return count
```

```python
assert(part_1(example) == 3)
print(part_1(input))
```

```python
def part_2(input):
    intervals, _ = input
    intervals.sort()
    merged = [intervals[0]]
    for i in range(1, len(intervals)):
        start, end = merged[-1]
        a, b = intervals[i]
        if a <= end:
            merged[-1] = [start, max(end, b)]
        else:
            merged.append([a, b])
    total = 0
    for a, b in merged:
        total += b - a + 1
    return total
```

```python
assert(part_2(example) == 14)
print(part_2(input))
```
